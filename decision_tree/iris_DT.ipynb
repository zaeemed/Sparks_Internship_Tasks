# libraries

```python
import pandas as pd
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import tree
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
```

# loading dataset

```python
iris = datasets.load_iris()
class_names = iris.target_names
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df['target'] = iris.target
iris_df['target'] = iris_df['target'].replace([0,1,2],['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
```

```python
iris_df.head()
```

# initialization of training and testing data

```python
X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=15)

accuracy_list = []
```

## min_samples_split

```python
clf1 = DecisionTreeClassifier(min_samples_split = 3)
clf1 = clf1.fit(X_train, y_train)
y_pred1 = clf1.predict(X_test)
accuracy = metrics.accuracy_score(y_test, y_pred1)
accuracy_list.append(f"Accuracy of decision tree made by min_samples_split is: {accuracy}")
```

## min_weight_fraction_leaf

```python
clf2 = DecisionTreeClassifier(min_weight_fraction_leaf = 0.3)
clf2 = clf2.fit(X_train, y_train)
y_pred2 = clf2.predict(X_test)
accuracy = metrics.accuracy_score(y_test, y_pred2)
accuracy_list.append(f"Accuracy of decision tree made by min_weight_fraction_leaf is: {accuracy}")
```

## max_features

```python
clf3 = DecisionTreeClassifier(max_features= 3)
clf3 = clf3.fit(X_train, y_train)
y_pred3 = clf3.predict(X_test)
accuracy = metrics.accuracy_score(y_test, y_pred3)
accuracy_list.append(f"Accuracy of decision tree made by max_features is: {accuracy}")
```

```python
for i in accuracy_list:
  print(i)
```

```python
feature_cols = ['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
```

## min_samples_split

```python
print(confusion_matrix(y_test, y_pred1))
print(classification_report(y_test, y_pred1))
```
